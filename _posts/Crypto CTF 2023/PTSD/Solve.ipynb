```python
import telnetlib
from random import randint
from Crypto.Util.number import *
host = "05.cr.yp.toc.tf" 
port = int(11137)

def read_multiple_lines(num):
    for i in range(num):
        print(tn.read_until(b"\n"))

temp_limit = [12, 12, 12, 16, 19, 22, 27, 30, 33, 36, 39, 42, 46, 50, 54, 58, 62, 66, 69, 73, 77, 81, 85, 89, 93, 93, 96, 99, 102]
temp_step = [0, 0, 0, 1, 1, 1, 13, 15, 19, 22, 25, 28, 30, 33, 37, 41, 45, 51, 55, 58, 61, 65, 70, 74, 78, 82, 82, 87, 88, 90]
request_list = []
for i in range(len(temp_limit)):
    nbit_limit = temp_limit[i]
    nbit_step = temp_step[i]
    for r in range(-2**nbit_limit, 0, 2**nbit_step):
        a = 3*r - 9*r^4
        b = 1 - 9 * r^3
        c = 9 * r^4
        if is_prime(b) and b.nbits() == i*10 + 20:
            print(b.nbits())
            request_list.append(f"{a},{b},{c}\n".encode())
            break

print("Done")
            
tn = telnetlib.Telnet(host, port)

read_multiple_lines(11)

for request in request_list:
    tn.write(request)
    read_multiple_lines(3)

tn.close()



```
